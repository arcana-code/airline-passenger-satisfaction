```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("airline_passenger_satisfaction.csv")
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df['Arrival Delay in Minutes']=df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].mean)
```

```python
categorical=['Gender','Customer Type','Type of Travel','Class','satisfaction']
for column in categorical:
    print(f"unique values in{column}are{df[column].unique()}")
```

```python
def c(x):
    if x=='neutral or dissatisfied':
        return 0
    else:
        return 1
```

```python
df[ 'satisfaction'] = df[ 'satisfaction'].apply(c)
```

```python
df.drop(columns=['Unnamed: 0', 'id'],inplace=True)
```

```python
df.head(2)
```

```python
sns.countplot(x='Customer Type',data=df)
plt.show()
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```

```python
x= df.drop(columns='satisfaction')
x
```

```python
y=df['satisfaction']## Target (dependent variable)
y
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
pd.DataFrame(x_scaled,columns = x.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5)
```

```python
x_train.size
```

```python
y_train.size
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=100,max_depth=10,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
c=confusion_matrix(y_test,y_pred)
c
```

```python
plt.figure(plt.figure(figsize=(5,4)))
sns.heatmap(c,annot=True,fmt=".1f")
```
